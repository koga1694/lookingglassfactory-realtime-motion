```python
def com_joint(point, a, b, c):
        joint = np.arctan((point[c][1]-point[b][1]) / (point[c][0] - point[b][0])) - np.arctan((point[a][1] - point[b][1]) / (point[a][0] - point[b][0]))
        
        return joint

def get_joint(point):
    joint1 = com_joint(point, 2, 0, 3)
    joint2 = com_joint(point, 0, 2, 1)
    joint3 = com_joint(point, 6, 7, 8)
    joint4 = com_joint(point, 9, 7, 10)
    joint5 = com_joint(point, 0, 7, 1)
    joint6 = com_joint(point, 1, 5, 8)
    joint7 = com_joint(point, 1, 10, 8)
    joint8 = com_joint(point, 13, 12, 14)
    joint9 = com_joint(point, 21, 22, 23)
    joint10 = com_joint(point, 6, 19, 23)
    
    joint_list = [joint1, joint2, joint3, joint4, joint5, joint6, joint7, joint8, joint9, joint10]
    
#     print(joint_list)
    return joint_list

def cos_joint(point, a, b, c):
    x = 0
    y = 1
    joint = np.arccos(
        (((point[c][x] - point[b][x])*(point[a][x] - point[b][x])) + ((point[c][y] - point[b][y])*(point[a][y] - point[b][y]))) / (np.sqrt((point[c][x] - point[b][x])**2 + (point[c][y] - point[b][y])**2)*np.sqrt((point[a][x] - point[b][x])**2 + (point[a][y] - point[b][y])**2))
    )
    
    return joint

def cos_get_joint(point):
    joint1 = cos_joint(point, 2, 0, 3)
    joint2 = cos_joint(point, 0, 2, 1)
    joint3 = cos_joint(point, 6, 7, 8)
    joint4 = cos_joint(point, 9, 7, 10)
    joint5 = cos_joint(point, 0, 7, 1)
    joint6 = cos_joint(point, 1, 5, 8)
    joint7 = cos_joint(point, 1, 10, 8)
    joint8 = cos_joint(point, 13, 12, 14)
    joint9 = cos_joint(point, 21, 22, 23)
    joint10 = cos_joint(point, 6, 19, 23)
    
    joint_list = [joint1, joint2, joint3, joint4, joint5, joint6, joint7, joint8, joint9, joint10]
    
    return joint_list

def get_vector(point, a, b, c):
    v1 = point[c] - point[b]
    v2 = point[a] - point[b]
    
    v_len1 = np.sqrt((v1**2).sum())
    v_len2 = np.sqrt((v2**2).sum())
    
    angle = np.arccos((v1*v2).sum() / (v_len1 * v_len2))
#     angle = np.degrees(angle)
    
    return angle
    

def vec_joint(point):
    joint1 = get_vector(point, 2, 0, 3)
    joint2 = get_vector(point, 0, 2, 1)
    joint3 = get_vector(point, 6, 7, 8)
    joint4 = get_vector(point, 9, 7, 10)
    joint5 = get_vector(point, 0, 7, 1)
    joint6 = get_vector(point, 1, 5, 8)
    joint7 = get_vector(point, 1, 10, 8)
    joint8 = get_vector(point, 13, 12, 14)
    joint9 = get_vector(point, 21, 22, 23)
    joint10 = get_vector(point, 6, 19, 23)
    
    joint_list = [joint1, joint2, joint3, joint4, joint5, joint6, joint7, joint8, joint9, joint10]
    
    return joint_list
    

point1 = np.array([0.4382759,0.41970551,0.0163717 ])
point2 = np.array([ 0.51665819,  0.55781358, -0.06201499])
point3 = np.array([ 0.57319468,  0.6209566 , -0.00924825])

v1 = point3 - point2
v2 = point1 - point2
print(v1,v2)
v_len1 = np.sqrt((v1**2).sum())
v_len2 = np.sqrt((v2**2).sum())

print(v_len1)
print(v_len2)




angle = np.arccos((v1*v2).sum() / (v_len1 * v_len2))
angle = np.degrees(angle)
angle
```

```python
# knn 모델 
from sklearn.neighbors import KNeighborsClassifier

emotion = {0:'AN', 1:'DI', 2 :'FE', 3:'HA', 4:'NE', 5:'SA', 6: 'SU'}  
cos_file = np.genfromtxt('cos_joint.csv', delimiter=',')
tan_file = np.genfromtxt('tan_joint.csv', delimiter=',')
k_file = np.genfromtxt('k_joint.csv', delimiter=',')
k_cos_file = np.genfromtxt('k_cos_joint.csv', delimiter=',')
vec_file = np.genfromtxt('vector_joint.csv', delimiter=',')

cos_angle = cos_file[:,:-1].astype(np.float32)
cos_label = cos_file[:, -1].astype(np.float32)

k_angle = k_file[:,:-1].astype(np.float32)
k_label = k_file[:, -1].astype(np.float32)

tan_angle = tan_file[:,:-1].astype(np.float32)
tan_label = tan_file[:, -1].astype(np.float32)

k_cos_label = k_cos_file[:, -1].astype(np.float32)
k_cos_angle = k_cos_file[:,:-1].astype(np.float32)

vec_label = vec_file[:, -1].astype(np.float32)
vec_angle = vec_file[:,:-1].astype(np.float32)


cos_knn_model = KNeighborsClassifier(n_neighbors=1, metric="minkowski",
                          weights = 'uniform', p = 2, leaf_size = 30)

tan_knn_model = KNeighborsClassifier(n_neighbors=1, metric="minkowski",
                          weights = 'uniform', p = 2, leaf_size = 30)

k_knn_model = KNeighborsClassifier(n_neighbors=1, metric="minkowski",
                          weights = 'uniform', p = 2, leaf_size = 30)
k_cos_knn_model = KNeighborsClassifier(n_neighbors=1, metric="minkowski",
                          weights = 'uniform', p = 2, leaf_size = 30)

vec_knn_model = KNeighborsClassifier(n_neighbors=1, metric="minkowski",
                          weights = 'uniform', p = 2, leaf_size = 30)
cos_knn_model.fit(cos_angle, cos_label)
tan_knn_model.fit(cos_angle, cos_label)
k_knn_model.fit(k_angle, k_label)
k_cos_knn_model.fit(k_cos_angle, k_cos_label)
vec_knn_model.fit(vec_angle, vec_label)
```
